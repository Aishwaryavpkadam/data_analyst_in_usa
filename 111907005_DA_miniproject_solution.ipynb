### All the data entries in the 'Salary Estimate' column is in string format with (Glassdoor est.) in every cell, so let us extract only the numbers from the datasets so that we could use it.

### Create 2 new columns that tell about the minimum and maximum estimate about the Salary.

```python
df['Min Estimate'] = df["Salary Estimate"].str.extractall(r"[$](\d+)").xs(0, level='match')
df['Max Estimate'] = df["Salary Estimate"].str.extractall(r"[$](\d+)").xs(1, level='match')
```

```python
df['Max Estimate'].fillna(-1,inplace=True)
df['Min Estimate'].fillna(-1,inplace=True)
df['Min Estimate']=df['Min Estimate'].astype(int)
df['Max Estimate']=df['Max Estimate'].astype(int)
```

```python
df.head(3)
```

### Remove the Location of the Company from the Company Name column

```python
df["Company Name"] = df["Company Name"].str.replace("(\n).*","")
```

### Add columns that state the Location of the State and the City

```python
df["Location State"] = df["Location"].agg(lambda x: x[-2:])
df["Location City"] = df["Location"].agg(lambda x: x[:-4])
```

```python
df.head(3)
```

### Add 2 columns that give the average salary estimation and the range of the salary.

```python
df['Salary Mean'] = (df["Min Estimate"] + df['Max Estimate'])/2
```

```python
df['Salary Range'] = df["Max Estimate"] + df['Min Estimate']
```

```python
fig,(ax0, ax1) = plt.subplots(nrows =1, ncols = 2, figsize = (20,4))
sns.histplot(df['Min Estimate'], color = 'purple', ax = ax0,kde = True)
sns.histplot(df['Max Estimate'], color = 'green', ax = ax1, kde = True)
ax0.set(xlabel = "Lower Estimate ($ k/year)", ylabel = 'Percentage')
ax0.set(title = "Lower Limit for the Salary Estimation")
ax1.set(xlabel = "Upper Estimate ($ k/year)", ylabel = 'Percentage')
ax1.set(title = "Upper Limit for the Salary Estimation")
sns.set_style("white")
plt.show()
```

#### Both the above graphs depict that the salary estimataes are approximately normally distributed. We can conlcude from this graph that majority of the people expect to have a similar salary which implies there is uniformity in the pay range. Few outliers in the graph implies that those individuals would be placed in the higher level of the hierrarchy and hence estimate to be a higher salary.

```python
fig,(ax0, ax1) = plt.subplots(nrows =1, ncols = 2, figsize = (20,4))
sns.histplot(df['Salary Mean'], color = 'red', ax = ax0)
sns.histplot(df['Salary Range'], color = 'teal', ax = ax1)
ax1.set(xlabel = 'Salary Range ($ k/year)', ylabel = 'Percentage')
ax1.set(title = 'Salary Range Distribution')
ax0.set(xlabel = 'Average Salary ($ k/year)', ylabel = 'Percentage')
ax0.set(title = 'Histogram for the Averge Salary Estimates')
plt.show()
```

### Display the Average/Mean Salaries of the top 10 sectors through the Violin Plot.
